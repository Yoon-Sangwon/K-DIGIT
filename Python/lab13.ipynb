```python
import pandas as pd

df = pd.DataFrame({'1분기' : [500, 690, 1100, 1500, 1990, 1020],
                   '2분기' : [450, 700, 1030, 1650, 2020, 1600],
                   '3분기' : [520, 820, 1200, 1700, 2300, 2200],
                   '4분기' : [610, 900, 1380, 1850, 2420, 2550]}, index = [2015, 2016, 2017, 2018, 2019, 2020])
display(df)
```

```python
from matplotlib import font_manager, rc
font_path = "data/THEdog.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(df)
title_font = {
    'fontsize': 20,
    'fontweight': 'bold',
    'color' : 'purple'
}
plt.title('시각화 과제1', fontdict = title_font, pad=20)
plt.legend(df.columns)
plt.savefig("output/hw1.png")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.T)
title_font = {
    'fontsize': 20,
    'fontweight': 'bold',
    'color' : 'red'
}
plt.title('시각화 과제2', fontdict = title_font, pad=20)
plt.legend(df.index)
plt.savefig("output/hw2.png")
plt.show()
```

```python
plt.figure(figsize=(10,6))
explode = [0.05, 0.05, 0.05, 0.05]
plt.pie(df.sum(), labels=df.columns, autopct='%.1f%%', explode=explode, shadow=True)
plt.savefig("output/hw3.png")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(df.T.sum(), labels=df.T.columns, autopct='%.1f%%')
plt.savefig("output/hw4.png")
plt.show()
```
