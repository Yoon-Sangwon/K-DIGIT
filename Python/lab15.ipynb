```python
import pandas as pd
import seaborn as sns
import numpy as np
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
                     'petal_length', 'petal_width']]
import random
random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
iris_x.head(10)
```

```python
iris_x.head(10).dropna(axis=0)
```

```python
iris_x.head(12).dropna(subset=['sepal_length', 'petal_width'], how='all', axis=0)
```

```python
iris_x.head(15).dropna(subset=['sepal_length', 'sepal_width'], how='any', axis=0)
```

```python
iris_x.head(10).fillna(0)
```

```python
iris_x.head(10).fillna(round(iris_x.mean(),1))
```

```python
iris_x.head(10).fillna(10)
```

```python
iris_x.head(10).fillna(method='ffill')
```

```python
log = pd.read_csv('./data/product_click_new.log', sep = ' ', header = None)
print(log.head(5))
print(log.info())
```

```python
log.columns = ['oldclicktime', 'clickpid']
print(log.head(5))
```

```python
log['oldclicktime'] = log['oldclicktime'].astype('object')
print(log.info())
```

```python
log['newclicktime'] = pd.to_datetime(log['oldclicktime'].astype('string'))
print(log.info())
print(log.head(5))
```

```python
log['year'] = log['newclicktime'].dt.year
log['month'] = log['newclicktime'].dt.month
log['day'] = log['newclicktime'].dt.day
log['hour'] = log['newclicktime'].dt.hour
log['minute'] = log['newclicktime'].dt.minute
print(log.info())
log.head(5)
```
