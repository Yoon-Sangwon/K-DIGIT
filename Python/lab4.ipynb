```python
import requests
from bs4 import BeautifulSoup
import re

title = []
summary = []
rating = []

for i in range(1, 11):
    req = requests.get('https://comic.naver.com/genre/bestChallenge.nhn?&page=' + str(i))
    req.encoding = 'utf-8'
    html = req.text
    bs = BeautifulSoup(html, 'html.parser')
    titles = bs.select('div.challengeInfo > h6 > a')
    summarys = bs.select('div.challengeInfo > div.summary')
    ratings = bs.select('div.challengeInfo > div.rating_type > strong')
    for content in titles:
        content=re.sub("[\n\t\s]", '', content.text)
        title.append(content)
    for content in summarys:
        summary.append(content.text)
    for content in ratings:
        rating.append(content.text)

print(title)
print(summary)
print(rating)
```
